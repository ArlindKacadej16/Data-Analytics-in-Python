```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6)) 
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6)) 
plt.plot(rev)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
GROSS_profit = rev + COGS
GROSS_profit
plt.figure(figsize=(15,6)) 
plt.plot(rev)
plt.show()
```
